```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Calories'], bins=30, kde=True)
plt.title('Distribution of Caloric Intake')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.axvline(x=500, color='r', linestyle='--', label='500 Calories')
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.barplot(x='Cholesterol', y='Item', data=df.sort_values('Cholesterol', ascending=False).head(20), palette='viridis')
plt.title('Top 20 Items with Highest Cholesterol Content')
plt.xlabel('Cholesterol (mg)')
plt.ylabel('Item')
plt.show()
```

```python
#Sugar Intake

plt.figure(figsize=(10, 6))
sns.histplot(df['Sugars'], bins=30, kde=True, color='orange')
plt.title('Distribution of Sugar Content')
plt.xlabel('Sugars (g)')
plt.ylabel('Frequency')
plt.axvline(x=10, color='r', linestyle='--', label='10g Sugars')
plt.legend()
plt.show()
```

```python
# Ensure proper data types
df['Calories'] = pd.to_numeric(df['Calories'], errors='coerce')
df['Total Fat'] = pd.to_numeric(df['Total Fat'], errors='coerce')
df['Cholesterol'] = pd.to_numeric(df['Cholesterol'], errors='coerce')
df['Saturated Fat'] = pd.to_numeric(df['Saturated Fat'], errors='coerce')
df['Carbohydrates'] = pd.to_numeric(df['Carbohydrates'], errors='coerce')
df['Protein'] = pd.to_numeric(df['Protein'], errors='coerce')
df['Sugars'] = pd.to_numeric(df['Sugars'], errors='coerce')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Calories'], bins=30, kde=True)
plt.title('Distribution of Caloric Intake')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.axvline(x=1880, color='r', linestyle='--', label='Outlier at 1880 Calories')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Carbohydrates'], bins=30, kde=True, color='green')
plt.title('Distribution of Carbohydrates')
plt.xlabel('Carbohydrates (g)')
plt.ylabel('Frequency')
plt.axvline(x=40, color='r', linestyle='--', label='Peak at 40g')
plt.legend()
plt.show()
```

```python
# Protein vs Caloric Intake
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Calories', y='Protein', data=df)
plt.axvline(x=200, color='r', linestyle='--', label='2000 Calories')
plt.axvline(x=500, color='b', linestyle='--', label='2500 Calories')
plt.axhline(y=56, color='g', linestyle='--', label='56g Protein')
plt.title('Caloric Intake vs Protein')
plt.xlabel('Calories')
plt.ylabel('Protein (g)')
plt.legend()
plt.show()
```

```python
#Fat vs Caloric Intake
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Calories', y='Total Fat', data=df)
plt.axvline(x=2000, color='r', linestyle='--', label='2000 Calories')
plt.axvline(x=2500, color='b', linestyle='--', label='2500 Calories')
plt.axhline(y=78, color='g', linestyle='--', label='78g Fat')
plt.title('Caloric Intake vs Total Fat')
plt.xlabel('Calories')
plt.ylabel('Total Fat (g)')
plt.legend()
plt.show()
```

```python
 # Sugar vs Caloric Intake
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Calories', y='Sugars', data=df)
plt.axvline(x=1000, color='r', linestyle='--', label='1000 Calories')
plt.axhline(y=25, color='g', linestyle='--', label='25g Sugars (Women)')
plt.axhline(y=36, color='b', linestyle='--', label='36g Sugars (Men)')
plt.title('Caloric Intake vs Sugars')
plt.xlabel('Calories')
plt.ylabel('Sugars (g)')
plt.legend()
plt.show()
```

```python
# Define nutritional limits for women
caloric_limit = 2500
protein_limit = 56
fat_limit = 78
sugar_limit = 25

# Filter items within the limits
suitable_items = df[
    (df['Calories'] <= caloric_limit) &
    (df['Protein'] >= protein_limit - 10) & # Allowing a bit of flexibility
    (df['Total Fat'] <= fat_limit) &
    (df['Sugars'] <= sugar_limit)
]

# Sort by Calories to find items within moderate calorie range
suitable_items = suitable_items.sort_values(by='Calories')

# Display the top 10 suitable items
suitable_items[['Item', 'Calories', 'Total Fat', 'Cholesterol', 'Protein', 'Sugars']].head(60)
```

```python
# Define nutritional limits for men
caloric_limit = 2500
protein_limit = 70
fat_limit = 90
sugar_limit = 150

# Filter items within the limits
suitable_items = df[
    (df['Calories'] <= caloric_limit) &
    (df['Protein'] >= protein_limit - 10) & # Allowing a bit of flexibility
    (df['Total Fat'] <= fat_limit) &
    (df['Sugars'] <= sugar_limit)
]

# Sort by Calories to find items within moderate calorie range
suitable_items = suitable_items.sort_values(by='Calories')

# Display the top 10 suitable items
suitable_items[['Item', 'Calories', 'Total Fat', 'Cholesterol', 'Protein', 'Sugars']].head()
```
